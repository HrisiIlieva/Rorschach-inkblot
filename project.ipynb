## Introduction: The Rorschach Inkblot and Mathematics

### What is a Rorschach Inkblot?
The Rorschach test, created by psychologist Hermann Rorschach in 1921, consists of symmetrical inkblots. While originally designed for psychological evaluation, these shapes represent a fascinating intersection of **order** and **chaos**. The mind perceives patterns (Pareidolia) in what is essentially a random distribution of ink.

## Mathematical Analysis

### 1. Geometric Model of the Inkblot

Each inkblot can be modeled as a subset of the plane

$$
B \subset \mathbb{R}^2
$$

where the points belong to the region covered with ink.

In practice, the inkblot is created by spilling ink and folding the paper, which produces mirror symmetry.

Mathematically, is defined a reflection operator $R$ acting on a position vector

$$
\mathbf{v} = (x,y)
$$

$$
R\mathbf{v} =
\begin{pmatrix}
-1 & 0 \\
0 & 1
\end{pmatrix}
\begin{pmatrix}
x \\
y
\end{pmatrix}
=
(-x,y)
$$

This means that the figure has axial symmetry with respect to the $y$-axis.

All standard 10 cards are approximately bilaterally symmetric, which is a key property of the test.

### 2. Symmetry Theory (Group Theory)

The inkblot is invariant under the action of the symmetry group

$$
G = \{e, r\}
$$

where $e$ is the identity and $r$ is the reflection.

This group is isomorphic to the cyclic group of order $2$:

$$
G \cong C_2
$$

Because

$$
(x,y) \in B \Rightarrow (-x,y) \in B
$$

the human brain's visual system automatically triggers "bilateral search" mode, often leading to the perception of faces or biological entities.

### 3. Fractal Geometry of the Inkblots

The boundaries of the inkblots are non-smooth and are best described using fractal dimension $D$.

If $N(\epsilon)$ is the number of boxes of size $\epsilon$ needed to cover the set $B$, then

$$
D =
\lim_{\epsilon \to 0}
\frac{\log N(\epsilon)}{\log(1/\epsilon)}
$$

Empirical studies show that Rorschach inkblots typically fall within

$$
1.1 < D < 1.3
$$

This specific range means the figure is more complex than a simple Euclidean line but less dense than a filled area, providing the optimal "perceptual ambiguity" required for the test.

### 4. Probabilistic Model of Perception

I model the interpretation of an inkblot using Bayesian probability.

Let

$$
S = \{s_1, s_2, ..., s_n\}
$$

be the set of possible interpretations.

The probability

$$
P(s_i \mid B)
$$

that an observer perceives object $s_i$ is given by

$$
P(s_i \mid B) =
\frac{P(B \mid s_i)P(s_i)}{P(B)}
$$

$P(s_i)$ is the prior probability based on the observer's experience.

$P(B \mid s_i)$ is the likelihood, representing how well the inkblot $B$ matches the mental template of $s_i$.

The symmetry of $B$ significantly increases $P(B \mid s_{\text{biological}})$ because most living organisms exhibit bilateral symmetry.

### 6. Topological Analysis

From a topological perspective, the inkblot is a connected set characterized by the Euler characteristic

$$
\chi = k - h
$$

where

- $k$ is the number of connected components  
- $h$ is the number of holes.

Visual complexity is further quantified by Shannon entropy

$$
H = -\sum p_i \log p_i
$$

A moderate $H$ is essential; if entropy is too low, the image is too obvious; if too high, it is perceived as meaningless noise.

### 7. Information Entropy

We can measure visual complexity using entropy

$$
H = -\sum p_i \log p_i
$$

where $p_i$ is the probability of different local structures.

The goal of the inkblot design is to maintain moderate entropy.

If it is too low → the figure becomes obvious.

If it is too high → the image looks like noise.
