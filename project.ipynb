## Mathematical Analysis

### 1. Geometric Model of the Inkblot

Each inkblot can be modeled as a subset of the plane

$$
B \subset \mathbb{R}^2
$$

where the points belong to the region covered with ink.

In practice, the inkblot is created by spilling ink and folding the paper, which produces mirror symmetry.

Mathematically, is defined a reflection operator $R$ acting on a position vector

$$
\mathbf{v} = (x,y)
$$

$$
R\mathbf{v} =
\begin{pmatrix}
-1 & 0 \\
0 & 1
\end{pmatrix}
\begin{pmatrix}
x \\
y
\end{pmatrix}
=
(-x,y)
$$

This means that the figure has axial symmetry with respect to the $y$-axis.

All standard 10 cards are approximately bilaterally symmetric, which is a key property of the test.

### 2. Symmetry Theory (Group Theory)

The inkblot is invariant under the action of the symmetry group

$$
G = \{e, r\}
$$

where $e$ is the identity and $r$ is the reflection.

This group is isomorphic to the cyclic group of order $2$:

$$
G \cong C_2
$$

Because

$$
(x,y) \in B \Rightarrow (-x,y) \in B
$$

the human brain's visual system automatically triggers "bilateral search" mode, often leading to the perception of faces or biological entities.

### 3. Fractal Geometry of the Inkblots

The boundaries of the inkblots are non-smooth and are best described using fractal dimension $D$.

If $N(\epsilon)$ is the number of boxes of size $\epsilon$ needed to cover the set $B$, then

$$
D =
\lim_{\epsilon \to 0}
\frac{\log N(\epsilon)}{\log(1/\epsilon)}
$$

Empirical studies show that Rorschach inkblots typically fall within

$$
1.2 < D < 1.6
$$

This specific range means the figure is more complex than a simple Euclidean line but less dense than a filled area, providing the optimal "perceptual ambiguity" required for the test.

### 4. Probabilistic Model of Perception

I model the interpretation of an inkblot using Bayesian probability.

Let

$$
S = \{s_1, s_2, ..., s_n\}
$$

be the set of possible interpretations.

The probability

$$
P(s_i \mid B)
$$

that an observer perceives object $s_i$ is given by

$$
P(s_i \mid B) =
\frac{P(B \mid s_i)P(s_i)}{P(B)}
$$

$P(s_i)$ is the prior probability based on the observer's experience.

$P(B \mid s_i)$ is the likelihood, representing how well the inkblot $B$ matches the mental template of $s_i$.

The symmetry of $B$ significantly increases $P(B \mid s_{\text{biological}})$ because most living organisms exhibit bilateral symmetry.

### 6. Topological Analysis

From a topological perspective, the inkblot is a connected set characterized by the Euler characteristic

$$
\chi = k - h
$$

where

- $k$ is the number of connected components  
- $h$ is the number of holes.

Visual complexity is further quantified by Shannon entropy

$$
H = -\sum p_i \log p_i
$$

A moderate $H$ is essential; if entropy is too low, the image is too obvious; if too high, it is perceived as meaningless noise.

### 7. Information Entropy

We can measure visual complexity using entropy

$$
H = -\sum p_i \log p_i
$$

where $p_i$ is the probability of different local structures.

The goal of the inkblot design is to maintain moderate entropy.

If it is too low → the figure becomes obvious.

If it is too high → the image looks like noise.

## Generating an Inkblot with Python

Mathematically, I will generate an inkblot similar to those in the Rorschach test using Python and random processes.

The idea is to simulate how ink spreads and then forms mirror symmetry.

### 1. Mathematical Idea

I will use three main steps.

#### 1️ Random Field

I generate a random function

$$
f(x,y)
$$

which describes the density of the ink.

$$
f(x,y) \sim N(0,1)
$$

(normal distribution).

#### 2️ Threshold Function

I transform the random field into a black inkblot using a threshold:

$$
B(x,y) =
\begin{cases}
1 & f(x,y) > t \\
0 & f(x,y) \le t
\end{cases}
$$

where $t$ is the threshold.

#### 3️ Mirror Symmetry

To obtain the characteristic shape of the test:

$$
B(x,y) = B(-x,y)
$$

That is, the left half is copied onto the right half.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import laplace, gaussian_filter
```

```python
# grid size
n = 300

# random field
noise = np.random.normal(0, 1, (n, n//2))

# threshold
threshold = 0.7
blob = noise > threshold

# mirror symmetry
ink = np.concatenate((blob, np.fliplr(blob)), axis=1)

# visualization
plt.imshow(ink, cmap="Greys")
plt.axis("off")
plt.show()
```

This code:

- creates a random field
- generates a black-and-white inkblot
- adds mirror symmetry
- displays the result

Since real inkblots are more complex because the ink spreads,
I will extend the previous code by adding a diffusion model that simulates the spreading of ink from a random initial point, described by the equation

$$
\frac{\partial u}{\partial t} = D \nabla^2 u
$$

where

- $u(x,y,t)$ — ink concentration
- $D$ — diffusion coefficient
- $\nabla^2$ — Laplacian (spreading to neighboring points)

After that I will apply a threshold and mirror symmetry, similar to the Rorschach test inkblots.

```python
# grid size
n = 300

# initial ink concentration (several random points)
u = np.zeros((n, n//2))

num_seeds = 100
for _ in range(num_seeds):
    x = np.random.randint(0, n)
    y = np.random.randint(0, n//2)
    u[x, y] = 1.0

# diffusion parameters
D = 1.4
dt = 0.1
steps = 200

# diffusion simulation
for _ in range(steps):
    u += D * laplace(u) * dt

# normalization
u = (u - u.min()) / (u.max() - u.min())

# threshold
threshold = 0.2
blob = u > threshold

# mirror symmetry
ink = np.concatenate((blob, np.fliplr(blob)), axis=1)

# visualization
plt.figure(figsize=(6,6))
plt.imshow(ink, cmap="binary")
plt.axis("off")
plt.show()
```

What this model does

1️ Initial points  
Random "drops of ink".

2️ Diffusion  
The Laplacian simulates the spreading of ink through the paper.

3️ Threshold  
Determines the boundary between black ink and white paper.

4️ Symmetry  
The left half is copied onto the right half.

To obtain natural boundaries, I will extend the previous model by adding fractal enhancement using Perlin-like noise (fractal noise).

This makes the boundaries of the inkblot more organic and closer to real ink shapes.

The idea is:

$$
u_{final}(x,y) = u_{diffusion}(x,y) + \alpha \cdot F(x,y)
$$

where

- $u_{diffusion}$ — result of the diffusion
- $F(x,y)$ — fractal noise
- $\alpha$ — strength of the fractal influence

```python
# grid size
n = 300

# initial ink concentration
u = np.zeros((n, n//2))

num_seeds = 100
for _ in range(num_seeds):
    x = np.random.randint(0, n)
    y = np.random.randint(0, n//2)
    u[x, y] = 1.0

# diffusion parameters
D = 1.4
dt = 0.1
steps = 200

# diffusion simulation
for _ in range(steps):
    u += D * laplace(u) * dt

# normalization
u = (u - u.min()) / (u.max() - u.min())

# fractal noise

def fractal_noise(shape, octaves=4):
    noise = np.zeros(shape)
    amplitude = 1
    total_amp = 0

    for i in range(octaves):
        random_layer = np.random.rand(*shape)
        smooth_layer = gaussian_filter(random_layer, sigma=2**i)

        noise += amplitude * smooth_layer
        total_amp += amplitude
        amplitude /= 2

    return noise / total_amp

noise = fractal_noise(u.shape)

# add fractal noise
alpha = 0.4
u = u + alpha * noise

# normalization
u = (u - u.min()) / (u.max() - u.min())

# threshold
threshold = 0.2
blob = u > threshold

# mirror symmetry
ink = np.concatenate((blob, np.fliplr(blob)), axis=1)

# visualization
plt.figure(figsize=(6,6))
plt.imshow(ink, cmap="binary")
plt.axis("off")
plt.show()
```

What was added mathematically

1️ Fractal noise

Is created a sum of several "levels" of noise:

$$
F(x,y) = \sum_{i=0}^{k} a_i N_i(x,y)
$$

where

- $N_i$ is smoothed noise
- $a_i = 2^{-i}$

️2 Multi-scale structure

This creates details at different scales — characteristic of fractals.

3️ More realistic boundaries

Real inkblots have fractal dimensions approximately

$$
1.2 < D < 1.6
$$

therefore the fractal noise makes the result more natural.
