### Project Overview
The goal of this project is to explore the intersection of psychology and mathematics. 
By using Python, we will simulate the creation of inkblots through axial symmetry 
and statistical noise. This demonstrates how complex visual patterns can be 
reduced to simple linear algebra operations.

## Mathematical Model: Reflection Symmetry
A key characteristic of Rorschach blots is their bilateral symmetry. 
To achieve this, we use a reflection transformation. 

If we have a point $(x, y)$, its reflection across the vertical axis (y-axis) results in:
$$P'(x', y') = (-x, y)$$

In matrix notation, this linear transformation is expressed as:
$$\begin{pmatrix} x' \\ y' \end{pmatrix} = \begin{pmatrix} -1 & 0 \\ 0 & 1 \end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix}$$

By applying this to our generated noise, we create the mirror effect.
